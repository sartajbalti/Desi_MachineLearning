# Multiple Linear Regression 
> - More than two Variables 
> - One Dependent and more than one are independent variables

#### Step-1 Import data or data loading 

```python
# Importing data 
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
```

```python
# Importing Data 
df = pd.read_csv("salary1.csv")
df.head()
```

```python
X = df[['age','distance','YearsExperience']]
y = df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
a = X_train['age']
b = X_train['distance']
c = X_train['YearsExperience']
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[31,77,1.1]])
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
r = model.predict(X_train)
r
```

```python
# Visualization 
import matplotlib.pyplot as plt 
plt.scatter(a,y_train)
plt.scatter(b,y_train)
plt.scatter(c,y_train)
plt.plot(X_train,r,color='red')
plt.xlabel("Age || Distance || Years of Experience")
plt.ylabel("Salary")
plt.title("Salary visualization ")
```
